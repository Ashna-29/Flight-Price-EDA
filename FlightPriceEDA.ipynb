Changing the type of Date of Journey which is currently "str" to "int".

```python
## Checking data type of the Date of Journey column
df['Date_of_Journey'].dtype
```

Currently this column is "str" Datatype.

```python
## Converting the Date_of_Journey to Datetime format
df["Date_of_Journey"] = pd.to_datetime(df["Date_of_Journey"])
```

```python
## Again checking the data type of the Date of Journey column
df["Date_of_Journey"].dtype
```

Now this column is in datetime format.

Extracting Date and Month.


```python
df["Journey_day"] = df["Date_of_Journey"].dt.day
df["Journey_month"] = df["Date_of_Journey"].dt.month
df["Journey_year"] = df["Date_of_Journey"].dt.year
```

```python
df.head()
```

Now there is three seperate columns namely "Journey_day", "Journey_month" and "Journey_year".
Now we can delete the original Date_of_Journey columns

```python
## Dropping the Date_of_Journey column as we have extracted the day, month and year from it.
df.drop(columns=["Date_of_Journey"], inplace=True)
```

```python
df["Dep_Time"].dtype
```

Changing datatype of Dep_Time first to datetime format then to int.

```python
## Converting the Dep_Time column to datetime format
df["Dep_Time"] = pd.to_datetime(df["Dep_Time"], format="%H:%M")
```

```python
df.head()
```

```python
## Extracting the hour and minute from the Dep_Time column
df["Dep_hour"] = df["Dep_Time"].dt.hour
df["Dep_minute"] = df["Dep_Time"].dt.minute
```

```python
df.head()
```

Now we have Dep_Time in int format we can now drop the original Dep_Time column

```python
## Dropping the Dep_Time column as we have extracted the hour and minute from it.
df.drop(columns=['Dep_Time'], inplace=True)
```

```python
df.head()
```

Cleaning Arrival Time.

```python
## Checking the data type of the Arrival_Time column
df["Arrival_Time"].dtype
```

Right now the Arrival_time is in "str" instead of "int", we will change it into "int".

```python

df["Arrival_Time"] = df["Arrival_Time"].str.split().str[0]

df["Arrival_Time"].head(10)
```

```python
df["Arrival_Time"] = pd.to_datetime(df["Arrival_Time"], format="%H:%M")
```

```python
df["Arrival_Hour"] = df["Arrival_Time"].dt.hour
df["Arrival_Minute"] = df["Arrival_Time"].dt.minute
```

```python
df.head()
```

Now we have extracted arrival time we can drop the original column.

```python
## Dropping the Arrival_Time column as we have extracted the hour and minute from it.
df.drop(columns=['Arrival_Time'], inplace=True)
```

Cleaning Durations

```python
df['Duration'].dtype
```

Right now Duration is in "str" dataype.


```python
df["Duration"].head(10)
```

```python
df["Duration"] = df["Duration"].apply(
    lambda x: x if "h" in x else "0h " + x ## Lambda function to add 0h to the duration if it does not contain hours.
)

df["Duration"] = df["Duration"].apply(
    lambda x: x if "m" in x else x + " 0m" ## Lambda function to add 0m to the duration if it does not contain minutes.
)
```

```python
df["Duration"].sample(10)
```

Now all the values have both h and m in the Duration.

Now extracting hours and minute seperately.


```python
df["Duration_Hours"] = df["Duration"].str.extract(r"(\d+)h").astype(int)
df["Duration_Minutes"] = df["Duration"].str.extract(r"(\d+)m").astype(int)
```

```python
df.head(10)
```

Converting duration hours anf duration minutes both into minutes.

```python
df["Duration_Total_Minutes"] = (
    df["Duration_Hours"] * 60 +
    df["Duration_Minutes"]
)
```

```python
df[["Duration", "Duration_Total_Minutes"]].head(10)
```

```python
df.drop(columns=["Duration_Hours", "Duration_Minutes"], inplace=True)
```

```python
df.head()
```

Now we will clean the Total stops.

```python
df["Total_Stops"].unique()
```

Now we will Map all the unique words.

```python
df["Total_Stops"] = df["Total_Stops"].map({
    "non-stop": 0,
    "1 stop": 1,
    "2 stops": 2,
    "3 stops": 3,
    "4 stops": 4
})
```

Now we will do Bivariate Analysis.

```python
plt.figure(figsize=(14,6))
sns.boxplot(x="Airline", y="Price", data=df, color="Red")
plt.xticks(rotation=90)
plt.title("Airline vs Price")
plt.show()

```

Jet Airways Business has the highest median ticket price.
Jet Airways Business also shows the largest price variation (widest IQR and long whiskers).
Among economy airlines, Jet Airways has the highest median price and widest spread.
Most airlines contain high-price outliers, indicating some unusually expensive tickets.

```python
sns.boxplot(x="Source", y="Price", data=df, color="pink")
plt.title("Source vs Price")
plt.show()
```

Flights departing from Delhi generally have the highest median ticket prices.
Chennai has the lowest median airfare among the source cities.
Bangalore and Delhi contain several extremely expensive flights, resulting in numerous high-value outliers.
Ticket prices vary considerably across departure cities, suggesting that the source location influences airfare.

```python
sns.boxplot(x="Destination", y="Price", data=df, color="Blue")
plt.title("Destination vs Price")
plt.show()
```

Flights to New Delhi and Cochin have the highest median ticket prices.
Flights to Delhi, Kolkata, and Hyderabad generally have lower fares.
Several destinations contain high-price outliers, indicating that some flights are significantly more expensive than the typical fare.
Destination appears to be an important factor affecting ticket prices.



```python
sns.boxplot(x="Total_Stops", y="Price", data=df, color="pink")
plt.title("Total Stops vs Price")
plt.show()
```

Ticket prices generally increase as the number of stops increases.
Non-stop flights (0 stops) have the lowest median price.
Flights with one, two, and three stops are progressively more expensive on average.
The highest-priced flights are mostly found among flights with one or two stops, as shown by the numerous outliers.
There are very few observations with four stops, so conclusions for this category should be interpreted cautiously.

We will find Correlation

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True),
            annot=True,
            cmap="coolwarm")
plt.show()
```

Total_Stops has the strongest positive correlation with Price (0.60), indicating that flights with more stops tend to be more expensive.
Duration_Total_Minutes also shows a moderate positive correlation with Price (0.50), suggesting longer flights generally cost more.
Total_Stops and Duration_Total_Minutes are strongly correlated (0.73), meaning flights with more stops usually take longer.
Journey day, journey month, departure time, and arrival time have very weak correlations with ticket price, indicating they have little linear relationship with airfare.
No strong negative correlations are observed in the dataset.
